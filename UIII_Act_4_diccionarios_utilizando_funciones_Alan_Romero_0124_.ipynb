### Diccionario de guardia

```python
# Inicialización del diccionario vacío
diccionario_guardia = {}

# --- FUNCIONES ---

def listar_guardia():
    print("\n" + "="*145)
    print(f"{'Nombre':<15} | {'Empresa':<12} | {'Puesto':<18} | {'Equipo':<18} | {'Turno':<10} | {'Protocolos':<25} | {'Capacitación':<20}")
    print("-" * 145)

    # Uso de for clave, valor (key, value) para mostrar todos los datos
    for nombre, datos in diccionario_guardia.items():
        print(f"{nombre:<15} | {datos['empresa']:<12} | {datos['puesto']:<18} | {datos['equipo']:<18} | {datos['turno']:<10} | {datos['protocolos']:<25} | {datos['capacitacion']:<20}")
    print("="*145 + "\n")

def agregar_guardia():
    nombre = input("Nombre del guardia: ")
    empresa = input("Empresa de seguridad: ")
    puesto = input("Puesto de vigilancia: ")
    equipo = input("Equipo de comunicación: ")
    turno = input("Turno: ")
    protocolos = input("Protocolos de acceso: ")
    capacitacion = input("Capacitación: ")

    # Guardar datos en el diccionario
    diccionario_guardia[nombre] = {
        'empresa': empresa,
        'puesto': puesto,
        'equipo': equipo,
        'turno': turno,
        'protocolos': protocolos,
        'capacitacion': capacitacion
    }
    print("\n¡Registro exitoso!")
    listar_guardia()

def buscar_guardia():
    nombre = input("Ingrese el nombre del guardia a buscar: ")
    if nombre in diccionario_guardia:
        datos = diccionario_guardia[nombre]
        print(f"\n>>> Datos encontrados para: {nombre}")
        for clave, valor in datos.items():
            print(f"- {clave.upper()}: {valor}")
    else:
        print("\n[!] El guardia no se encuentra en el sistema.")
    listar_guardia()

def borrar_guardia():
    nombre = input("Ingrese el nombre del guardia a eliminar: ")
    if nombre in diccionario_guardia:
        del diccionario_guardia[nombre]
        print(f"\n[!] Registro de {nombre} eliminado correctamente.")
    else:
        print("\n[!] No se encontró a nadie con ese nombre.")
    listar_guardia()

def actualizar_guardia():
    nombre = input("Ingrese el nombre del guardia que desea actualizar: ")
    if nombre in diccionario_guardia:
        print(f"Modificando datos de {nombre}:")
        empresa = input("Nueva Empresa: ")
        puesto = input("Nuevo Puesto: ")
        equipo = input("Nuevo Equipo: ")
        turno = input("Nuevo Turno: ")
        protocolos = input("Nuevos Protocolos: ")
        capacitacion = input("Nueva Capacitación: ")

        diccionario_guardia[nombre] = {
            'empresa': empresa,
            'puesto': puesto,
            'equipo': equipo,
            'turno': turno,
            'protocolos': protocolos,
            'capacitacion': capacitacion
        }
        print("\n¡Datos actualizados!")
    else:
        print("\n[!] El guardia no existe.")
    listar_guardia()

# --- MENÚ PRINCIPAL ---
while True:
    print("      SISTEMA DE GESTIÓN: GUARDIA")
    print("1. Agregar guardia")
    print("2. Listar guardias (Tabla completa)")
    print("3. Buscar guardia")
    print("4. Borrar guardia")
    print("5. Actualizar guardia")
    print("6. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_guardia()
    elif opcion == "2":
        listar_guardia()
    elif opcion == "3":
        buscar_guardia()
    elif opcion == "4":
        borrar_guardia()
    elif opcion == "5":
        actualizar_guardia()
    elif opcion == "6":
        print("Cerrando sistema...")
        break
    else:
        print("Opción inválida.")
```
